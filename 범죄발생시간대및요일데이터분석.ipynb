```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc

# matplotlib   
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

###   ###
df = pd.read_csv('data_2019.csv',encoding='euc-kr')

###   ###

# 
# 1)    

#   ,  
data_1 = df.iloc[0:8, 2:10]

#   
data_1 = data_1.sum()

# 2)    

data_2 = df.iloc[0:8, 11:]
data_2 = data_2.sum()

# 

# 3)    
data_3 = df.iloc[8, 2:10]

# 4)    
data_4 = df.iloc[8, 11:]

# 

# 5)    
data_5 = df.iloc[9:17, 2:10]
data_5 = data_5.sum()

# 6)    
data_6 = df.iloc[9:17, 11:]
data_6 = data_6.sum()

###   ###

# 
plt.figure(figsize=(17,5))
plt.bar(data_1.index, data_1)
plt.title('  ')
plt.show

plt.figure(figsize=(17,5))
plt.bar(data_2.index, data_2)
plt.title('  ')
plt.show

# 
plt.figure(figsize=(17,5))
plt.bar(data_3.index, data_3)
plt.title('  ')
plt.show

plt.figure(figsize=(17,5))
plt.bar(data_4.index, data_4)
plt.title('  ')
plt.show

# 
plt.figure(figsize=(17,5))
plt.bar(data_5.index, data_5)
plt.title('  ')
plt.show

plt.figure(figsize=(17,5))
plt.bar(data_6.index, data_6)
plt.title('  ')
plt.show
```
