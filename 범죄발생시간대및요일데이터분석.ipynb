```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc

# matplotlib 한글 폰트 적용
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

### 데이터 불러오기 ###
df = pd.read_csv('data_2019.csv',encoding='euc-kr')

### 데이터 전처리 ###

# 강력범죄
# 1) 시간대별 강력범죄 발생횟수 데이터

# 필요한 데이터 행, 열 추출
data_1 = df.iloc[0:8, 2:10]

# 데이터 행 합계
data_1 = data_1.sum()

# 2) 요일별 강력범죄 발생횟수 데이터

data_2 = df.iloc[0:8, 11:]
data_2 = data_2.sum()

# 절도범죄

# 3) 시간대별 절도범죄 발생횟수 데이터
data_3 = df.iloc[8, 2:10]

# 4) 요일별 절도범죄 발생횟수 데이터
data_4 = df.iloc[8, 11:]

# 폭력범죄

# 5) 시간대별 폭력범죄 발생횟수 데이터
data_5 = df.iloc[9:17, 2:10]
data_5 = data_5.sum()

# 6) 요일별 폭력범죄 발생횟수 데이터
data_6 = df.iloc[9:17, 11:]
data_6 = data_6.sum()

### 데이터 시각화 ###

# 강력범죄
plt.figure(figsize=(17,5))
plt.bar(data_1.index, data_1)
plt.title('시간대별 강력범죄 발생횟수')
plt.show

plt.figure(figsize=(17,5))
plt.bar(data_2.index, data_2)
plt.title('요일별 강력범죄 발생횟수')
plt.show

# 절도범죄
plt.figure(figsize=(17,5))
plt.bar(data_3.index, data_3)
plt.title('시간대별 절도범죄 발생횟수')
plt.show

plt.figure(figsize=(17,5))
plt.bar(data_4.index, data_4)
plt.title('요일별 절도범죄 발생횟수')
plt.show

# 폭력범죄
plt.figure(figsize=(17,5))
plt.bar(data_5.index, data_5)
plt.title('시간대별 폭력범죄 발생횟수')
plt.show

plt.figure(figsize=(17,5))
plt.bar(data_6.index, data_6)
plt.title('요일별 폭력범죄 발생횟수')
plt.show
```
